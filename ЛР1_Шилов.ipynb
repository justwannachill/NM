18. Розв'яжіть систему лінійних рівнянь A-x=b, де А матриця коефіцієнтів, в вектор вільних членів.

```python
a = np.array([[1, 3, -4], [0, 2, -3], [-2, 2, 5]])
b= np.array([5, 3, 4])
x= np.linalg.solve(a, b)
print('Вбудована функція', x.round(1))

inv_a= np.linalg.inv(a)
x1 = inv_a.dot(b)
print('Матричний метод', x1.round(1))
```
