case - унікальний ідентифікатор кожного опосума в наборі. Просто порядковий номер.
site - місце відлову опосума. Це може бути географічна точка (наприклад, ліс, узлісся, околиця поселення), код або назва району в
Австралії чи Новій Гвінеї.

Рор - популяція, до якої належить тварина. Наприклад, може позначати континентальну та острівну групу (що часто
використовується у дослідженнях D. maculatus - австралійського опосума).
sex - стать опосума: "М" для самців, "F" для самок. Часто використовується для дослідження статевого диморфізму (різниці в
розмірах).
аде - вік, або ж ступінь дорослості (наприклад, дорослий/молодий). У деяких датасетах це може бути умовна категорія.
Морфометричні ознаки: hdlngth (head length) - довжина голови, важлива для аналізу розмірів черепа.
skullw (skull width) - ширина черепа. Часто корелює з масою тіла і використовується для видової класифікації.
totingth (total length) - загальна довжина опосума, включає і тулуб, і хвіст.
taill (tail length) - довжина хвоста. У опосумів хвіст грає важливу роль, зокрема в маневруванні в середовищі, тому цей параметр
показовий.

footlgth (foot length) - довжина задньої лапи. Зазвичай використовується для оцінки рухових можливостей.
earconch (ear conch length) - довжина вушної раковини. Цей показник допомагає у вивченні терморегуляції, бо у тварин з
більшими вухами ефективніше теплообмін.
еуe (eye diameter) - діаметр або розмір ока. У нічних тварин, як опосуми, часто великі очі - це адаптація до умов низького
освітлення.

chest - ширина або обхват грудей. Показує загальний розвиток тіла.
belly - ширина або обхват черевної частини тіла. Може відображати статеві, вікові або популяційні відмінності.

```python
from sklearn.decomposition import PCA

import matplotlib.pyplot as plt
import seaborn as sns
from kneed import KneeLocator

from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
from scipy.cluster.hierarchy import dendrogram

from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

X = df.select_dtypes(include=['number'])

X_scaled = StandardScaler().fit_transform(X)

pca = PCA(random_state=42).fit(X_scaled)
pve = pca.explained_variance_ratio_

print(pve)
```

```python
sns.set_theme()

kneedle = KneeLocator(
x=range(1, len(pve) + 1),
y=pve,
curve='convex',
direction='decreasing'
)

kneedle.plot_knee()
plt.title(f'Knee Point at {kneedle.elbow + 1}')
plt.xlabel("Number of components")
plt.ylabel("Explained Variance Ratio")
plt.grid(True)
plt.show()
```

```python
n_components = kneedle.elbow

ax = sns.lineplot(x=range(len(pve)), y=np.cumsum(pve))

ax.axvline(
    x=n_components,
    c='black',
    linestyle='--',
    linewidth=0.75
)

ax.axhline(
    y=np.cumsum(pve)[n_components],
    c='black',
    linestyle='--',
    linewidth=0.75
)

ax.set(
    xlabel='number of components',
    ylabel='cumulative explained variance'
)

plt.show()
```

```python
def plot_dendrogram(model, **kwargs):

    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)

    for i, merge in enumerate(model.children_):
        current_count = 0

        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1
            else:
                current_count += counts[child_idx - n_samples]

        counts[i] = current_count

    linkage_matrix = np.column_stack(
        [model.children_, model.distances_, counts]
    ).astype(float)

    dendrogram(linkage_matrix, **kwargs)
```

```python
model_agg = AgglomerativeClustering(
distance_threshold=0,
n_clusters=None)

model_agg = model_agg.fit(X)
```

```python
plot_dendrogram(model_agg, truncate_mode='level', p=3)
plt.title('Hierarchical Clustering Dendrogram' )
plt.xlabel('Number of points in node')

plt.show()
```

```python

X = df.select_dtypes(include=['number'])

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

model = KMeans(random_state=42)

visualizer = KElbowVisualizer(
    model,
    k=(2, 10),
    timings=False
)

visualizer.fit(X_scaled)
visualizer.show()

k_best = visualizer.elbow_value_

if k_best is None:
    k_best = 3

print("Best k:", k_best)

kmeans = KMeans(n_clusters=k_best, random_state=42)
agg = AgglomerativeClustering(n_clusters=k_best)

labels_kmeans = kmeans.fit_predict(X_scaled)
labels_agg = agg.fit_predict(X_scaled)

labels_kmn = pd.Series(labels_kmeans, name='k-means')
labels_agg = pd.Series(labels_agg, name='h-clust')

print(pd.crosstab(labels_kmn, labels_agg))

plt.figure(figsize=(10, 6))

sns.scatterplot(
    x=X_scaled[:, 0],
    y=X_scaled[:, 1],
    hue=labels_kmeans,
    palette='viridis'
)

plt.title('KMeans Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)

plt.show()
```
